### 2. Analysis Objective --> Rat Trading Detection

#### Case Introduction -- What is rat trading?

It mainly refers to the public fund that is operated by some people at first, usually investment directors, fund managers, etc. These ones first use personal funds to buy a stock at a low level, use public funds to raise the stock price, and then sell it at a high level for a personal account to make a profit.


## 3. Methodology --> How to Quantify Rat Trading

Feature 1: New Address --> Transaction numbers < 20

Feature 2: Invest Heavily Before the New --> Max Balance > 1,000 USD

Feature 3: Sell all Tokens After the New --> Token Balance ~ 0 USD

Feature 4: Short Time Interval Between Buy and Sell --> Time Interval < 30 days

```python
# Set Query Format
module = 'account'
action = 'txlist'
address = '0xD33526068D116cE69F19A9ee46F0bd304F21A51f'
startblock = 16427691 # Jan-16 2023
endblock = 16437100 # Just before the peak
sort = 'asc'


parameters = []
q_module = "https://api.etherscan.io/api?module={}".format(module); parameters.append(q_module)
q_action = "&action={}".format(action); parameters.append(q_action)
q_address = "&address={}".format(address); parameters.append(q_address)
q_startblock = "&startblock={}".format(startblock); parameters.append(q_startblock)
q_endblock = "&endblock={}".format(endblock); parameters.append(q_endblock)
q_sort = "sort={}".format(sort); parameters.append(q_sort)
q_offset = "offset={}".format(10000); parameters.append(q_offset)
q_apikey = "&apikey={}".format(apikey); parameters.append(q_apikey)


# Get the query URL
addr_list = []
url = ""
for p in parameters: url+=p
response = requests.get(url)
res = json.loads(response.text)['result']
for i in res:
  if i['to'] == address.lower():
    addr_list.append(i['from'])

addr_list = list(set(addr_list))
print('Select : {} Addresses'.format(len(addr_list)))
```

### 5.2 Check address transaction number

```python
# Calculate Addresses' transaction numbers before end_block
module = 'proxy'
action = 'eth_getTransactionCount'

# Set Query Format
parameters = []
pre_fix_url = ""
q_module = "https://api.etherscan.io/api?module={}".format(module); parameters.append(q_module)
q_action = "&action={}".format(action); parameters.append(q_action)
q_apikey = "&apikey={}".format(apikey); parameters.append(q_apikey)
q_blocknumber = "&blocknumber={}".format(endblock); parameters.append(q_blocknumber)
for p in parameters: pre_fix_url += p


# Query
import random
random.seed(2)
addr_list = random.sample(addr_list, k=200) # Randomly pick 200 addresses for demo
address_Tx_count = {}
for idx, addr_item in enumerate(addr_list): # delete reduplicative addresses
  if idx%10 == 0: print(idx, len(set(addr_list)))

  # Get the query URL
  url = pre_fix_url + "&address={}".format(addr_item)
  response = requests.get(url)
  # time.sleep(0.5)
  tx_Num = int(json.loads(response.text)['result'], 16)
  address_Tx_count[addr_item] = tx_Num

with open('./address_Tx_count.json', 'w') as f:
  json.dump(address_Tx_count, f)
```

### 5.3 Check balance before news to collect suspicious addresses

```python
def get_max_balance(address, contract_address, start_block, end_block, apikey, obsv_start_point=None):
    parameters = []
    url = ""
    q_module = "https://api.etherscan.io/api?module=account"; parameters.append(q_module)
    q_action = "&action=tokentx"; parameters.append(q_action)
    q_contract_address = "&contractaddress={}".format(contract_address); parameters.append(q_contract_address)
    q_address = "&address={}".format(address); parameters.append(q_address)
    q_start_block = "&startblock={}".format(start_block); parameters.append(q_start_block)
    q_end_block = "&endblock={}".format(end_block); parameters.append(q_end_block)
    q_page = "&page={}".format(1); parameters.append(q_page)
    q_offset = "&offset={}".format(1000); parameters.append(q_offset)
    q_sort = "sort={}".format('asc'); parameters.append(q_sort)
    q_apikey = "&apikey={}".format(apikey); parameters.append(q_apikey)
    for p in parameters: url += p


    response = requests.get(url)
    data = json.loads(response.text)['result']
    # print(data)
    max_balance = 0
    current_balance = 0

    for tx in data:
        if tx['to'].lower() == address.lower():
            current_balance += int(tx['value'])
        if tx['from'].lower() == address.lower():
            current_balance -= int(tx['value'])

        if obsv_start_point == None:
          max_balance = max(max_balance, current_balance)
        else:
          if int(tx['blockNumber']) > obsv_start_point:
            max_balance = max(max_balance, current_balance)

    return max_balance
```

```python
import time

# Calculate Addresses' Balance
start_block = 0
end_block = 16437100
RPL_price = 32.0
addr_bf_max_balance = {}

# Query
suspicious_addrs_1 = []
contract_addr = '0xD33526068D116cE69F19A9ee46F0bd304F21A51f'

for addr_item in set(suspicious_addrs):
  # Find the maximum balance
  max_balance = get_max_balance(addr_item, contract_addr, start_block, end_block, apikey)
  balance_in_USD = round(max_balance/1e18 * RPL_price, 2)
  time.sleep(0.5)
  addr_bf_max_balance[addr_item] = balance_in_USD

  # Collect suspicious addresses
  if balance_in_USD > 1000:
    print(addr_item, balance_in_USD)
    suspicious_addrs_1.append(addr_item)
```

### 5.4 Check balance after the news to filter suspicious addresses

```python
import time

# Calculate Addresses' Balance -- Block 16427690
start_block = 0
end_block = 16440100 # 16437100
RPL_price = 32.0
addr_af_max_balance = {}
suspicious_addrs_2 = []

# Query
contract_addr = '0xD33526068D116cE69F19A9ee46F0bd304F21A51f'

for addr_item in set(suspicious_addrs):
  # Find the maximum balance
  max_balance = get_max_balance(addr_item, contract_addr, start_block, end_block, apikey, obsv_start_point=16440000)
  balance_in_USD = round(max_balance/1e18 * RPL_price, 2)
  addr_af_max_balance[addr_item] = balance_in_USD
  time.sleep(0.5)

  # Collect suspicious addresses
  if balance_in_USD < 100:
    print(addr_item, balance_in_USD)
    suspicious_addrs_2.append(addr_item)
```

### 5.5 Check active period duration

```python
# Set Query Format
module = 'account'
action = 'txlist'
page = 1
offset = 0
sort = 'asc'


parameters = []
q_module = "https://api.etherscan.io/api?module={}".format(module); parameters.append(q_module)
q_action = "&action={}".format(action); parameters.append(q_action)
q_page = "&page={}".format(page); parameters.append(q_page)
q_offset = "&offset={}".format(offset); parameters.append(q_offset)
q_sort = "&sort={}".format(sort); parameters.append(q_sort)
q_apikey = "&apikey={}".format(apikey); parameters.append(q_apikey)
pre_fix_url = ""
for p in parameters: pre_fix_url+=p

suspicious_addrs_3 = []
addr_active_period = {}
for addr_item in suspicious_addrs:
  # Get the query URL
  addr_list = []
  url = pre_fix_url + "&address={}".format(addr_item)
  response = requests.get(url)
  res = json.loads(response.text)['result']
  time_span = (int(res[-1]['timeStamp']) - int(res[0]['timeStamp']))/3600./24
  addr_active_period[addr_item] = time_span

  print(addr_item, time_span)
  if time_span < 30: suspicious_addrs_3.append(addr_item)
```

```python
intersection = list(set(suspicious_addrs_1).intersection(suspicious_addrs_2, suspicious_addrs_3))
for i in intersection: print(i)
```

```python
from sklearn import tree
import numpy as np
from matplotlib import pyplot as plt

posi_label = []
posi_feature = []

nega_label = []
nega_feature = []

# Split Positive and Negative
for addr in suspicious_addrs:
  # Get Feature Vector
  feat_1 = address_Tx_count[addr]
  feat_2 = addr_bf_max_balance[addr]
  feat_3 = addr_af_max_balance[addr]
  feat_4 = addr_active_period[addr]

  if addr in intersection:
    posi_label.append(1)
    posi_feature.append([feat_1, feat_2, feat_3, feat_4])

  else:
    nega_label.append(0)
    nega_feature.append([feat_1, feat_2, feat_3, feat_4])


# Split Dataset
train_label = posi_label[:int(0.8*len(posi_label))] + nega_label[:int(0.8*len(nega_label))]
train_feature = posi_feature[:int(0.8*len(posi_label))] + nega_feature[:int(0.8*len(nega_label))]

test_label = posi_label[int(0.8*len(posi_label)):] + nega_label[int(0.8*len(nega_label)):]
test_feature = posi_feature[int(0.8*len(posi_label)):] + nega_feature[int(0.8*len(nega_label)):]

# Build and Train Decision Tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(train_feature, train_label)
pred = clf.predict(test_feature).tolist()

# Test
print('Prediction: {}'.format(pred))
print('Ground Truth: {}'.format(test_label))
print('Accuracy: {} %'.format(sum(np.array(test_label) == np.array(pred))/float(len(test_label))*100))
```
